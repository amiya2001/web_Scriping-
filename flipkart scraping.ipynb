## Here  My Problem Statement is  Scrap any E-Commerce Website product title, price and reviews 
 
 # Here I scrap the data of "Boat Headphones" from the E-commerce Site named "Flipkart"
   
   # Importing all Required Libraries

- Here first I import a Python Library named Beautifulsoup Which help me to Scrap the data from url and there are many other function which helps me in scraping the data.
- Then i Import requests module and after then I import  numpy and pandas Library.

```python
#Importing all required library
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
```

- After importing all the Required Library here I copy the link of a  product named "Boat Headphone "from the website named "flipkart" and store it in the variable named "url"

```python
#storing the url of the paticular product in a paticular website
url = "https://www.flipkart.com/search?q=boatheadphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

- Here I Exicute that the link of the website is work or not by the help of $"requests.get()"$ function and store the response in the variable named "Page" then print It

```python
#get the response that the url is work or not !
page =requests.get(url)
#printing the response
page 
```

- After getting Response I store all the data of the website in a variable named "html" by using the following code and print the data.

```python
#getting all the data from the url
html=page.content
#printing the data
html
```

- As you can see above the data present in the website is in HTML form and it's not arrange in a proper manner.
- So here I arrange the data in a proper HTML doc. Manner by writting the following code 

```python
#arrange the data in a proper HTML doc. manner
soup=bs(html,'html.parser')
#print the data
print(soup.prettify())
```

- After arrange the data I find all the required data such as "product name","product colour","product rating" etc. by the help of $find $_ $ all()$ function and store the data in different variables.

```python
#finding name of the product
b_names=soup.find_all("a",class_="s1Q9rs")
print(b_names)
print("\n"," ")

#finding colours of the product
b_colours=soup.find_all("div",class_="_3Djpdu")
print(b_colours)
print("\n"," ")

#finding ratings of the product
b_ratings=soup.find_all("div",class_="_3LWZlK")
print(b_ratings)
print("\n"," ")

#finding no. of ratings of the product
b_no_of_ratings=soup.find_all("span",class_="_2_R_DZ")
print(b_no_of_ratings)
print("\n"," ")

#finding original price of the product
b_original_price=soup.find_all("div",class_="_3I9_wc")
print(b_original_price)
print("\n"," ")

#finding discount  of the product
b_off_percentages=soup.find_all("div",class_="_3Ay6Sb")
print(b_off_percentages)
print("\n"," ")

#finding selling price of the product
b_selling_price=soup.find_all("div",class_="_30jeq3")
print(b_selling_price)
print("\n"," ")
```

- Here as you can see above all indivisual data is in HTML form so here I convert the data into a simple  list which conten only the data not HTML command and then print

```python
#declaring all empty list for storing data
names=[]
colour=[]
rating=[]
no_of_rating=[]
original_price=[]
off_percentage=[]
selling_price=[]


# by the help of loop  getting all the information about the product and storing in the Empty list
for nam in (b_names) :
    names.append(nam.get_text())
    
for col in (b_colours) :
    colour.append(col.get_text())
    
for rate in (b_ratings) :
    rating.append(rate.get_text())
    
for nrate in (b_no_of_ratings) :
    no_of_rating.append(nrate.get_text())
    
for oprice in (b_original_price) :
    original_price.append(oprice.get_text())
    
for off in (b_off_percentages) :
    off_percentage.append(off.get_text())
    
for price in (b_selling_price) :
    selling_price.append(price.get_text())
```

- Showing  all the list with information about product

```python
print(names,'\n','')
print(colour,'\n','')
print(rating,'\n','')
print(no_of_rating,'\n','')
print(original_price,'\n','')
print(off_percentage,'\n','')
print(selling_price,'\n','')
```

- Here i check that all the list have same length or not

```python
print(len(names))
print(len(colour))
print(len(rating))
print(len(no_of_rating))
print(len(original_price))
print(len(off_percentage))
print(len(selling_price))
```

- As you can see above that all the lis doesn't have same length.
- So adjust all list and make all list with same length

```python
#delete the extra rating,discount,selling price 
del rating[40:]
del off_percentage[40:]
del selling_price[40:]

#adding a null value in the original price by the help of numpy library
#because the price is missing
if len(original_price)==39:
    original_price.append(np.nan)  
```

- As you can see above all the list have same length. 
- Then here we are ready to convert all the data into a Tabular format .
- So here I convert all the data into a DataFrame by the help of Python Pandas library

```python
#converting all the data into DataFrame
df=pd.DataFrame({'Product Name ':names,'Product Colour':colour,'Product Rating':rating,
                 'No of Rating':no_of_rating,'Product Original Price':original_price,
                'Discount':off_percentage,'Product Selling Price':selling_price})
```

- Here I covert the DataFrame into Excel file and save it in the System by the name "flipkart scraping dataset".

```python
df.to_excel("flipkart scraping dataset.xlsx",index=True)
```
